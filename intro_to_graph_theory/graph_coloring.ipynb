# Graph Coloring

Graph coloring using a the greedy colouring algorithm.

## Packages

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.collections as collections
import networkx as nx
```

## Generate a graph

```python
# Create a random graph
np.random.seed(5)
G = nx.karate_club_graph()

# Draw graph
node_positions = nx.spring_layout(G)
fig, ax = plt.subplots(figsize=(8, 8))
plt.title('Graph')
nx.draw(G, pos = node_positions, node_color='lightgreen', ax=ax)
plt.show()
```

## Graph coloring

```python
def get_adjacencies(graph):
    adjacencies = {}
    for node_id, node_adjacencies in G.adjacency():
        adjacencies[node_id] = list(node_adjacencies.keys())       
    return adjacencies

def compute_groups(graph):
    adjencies = get_adjacencies(graph)
    num_vertices = len(graph)
    group_ids = [-1] * num_vertices

    for node, adjacencies in adjencies.items():
        # Get group ids from adjacencies
        adjacency_groups = []
        for adj in adjacencies:
            adjacency_groups.append(group_ids[adj])

        # Search for unassigned group id
        group_id = 0
        while group_id in adjacency_groups:
            group_id += 1

        group_ids[node] = group_id

    return np.asarray(group_ids)

group_ids = compute_groups(G)
```

## Show result

```python
 fig, ax = plt.subplots(figsize=(8, 8))
plt.title("Greedy Coloring Algorithm (%d colors)".format(len(group_ids)))
nx.draw(G, pos = node_positions, node_color=group_ids, ax=ax)
plt.show()
```
