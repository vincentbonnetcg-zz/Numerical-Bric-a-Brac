# Spectral Graph Theory

This notebooks introduces the concept of matrix analysis on graph matrices.
- Spectral layout
- Eigenvector centrality
- Spectral clustering


## Packages

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import networkx as nx
```

## Graph Creation

```python
NUM_NODES = 20
PROBABILITY = 0.3 # Probability for edge creation

np.random.seed(5)
G = nx.generators.erdos_renyi_graph(NUM_NODES, PROBABILITY, seed=np.random)

fig, ax = plt.subplots(figsize=(5, 5))
plt.title('Graph')
nx.draw(G, node_color='blue', node_size=10.0, with_labels = False,ax=ax)

ax.set_axis_on()
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
plt.show()
```

## Laplace Matrix

```python
A = nx.linalg.graphmatrix.adjacency_matrix(G).todense()
degree_per_vertex = [v for deg, v in G.degree()]
D = np.diagflat(degree_per_vertex)
L = nx.linalg.laplacianmatrix.laplacian_matrix(G).todense()
L = D - A
```

## Spectral Layout

```python
eigen_values, eigen_vectors = np.linalg.eigh(L)
ax = plt.subplot()
ax.plot(eigen_values)
plt.title('Eigen values in ascending order')
plt.show()

# get the two largest eigen values in descending order
# get their respective eigen vectors
largest_eigen_values = eigen_values[-1:-3:-1]
largest_eigen_vectors = eigen_vectors[-1:-3:-1]

# compute position

node_positions = nx.spectral_layout(G, weight=None)
```

#### Result

```python
fig, ax = plt.subplots(figsize=(10, 10))
plt.title('Spectral Layout ')
nx.draw(G, pos=node_positions, node_color='blue', node_size=1.0, with_labels = False,ax=ax)
ax.set_axis_on()
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
plt.show()
```
