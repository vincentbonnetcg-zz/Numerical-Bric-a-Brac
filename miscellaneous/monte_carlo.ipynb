# Integral Approximation
Monte Carlos is method to compute numerical integration using random samples over a domain. 

### Circle Area Approximation
We start with the approximation of the circle area. The analytic solution of the area of a circle is defined by $\pi r^{2}$ \
We avoid explicit for-loop by using vectorized numpy code to speed the computation.

```python
import matplotlib.pyplot as plt
import numpy as np
import random

RADIUS = 1.0
NUM_SAMPLES = 20000

# Create random samples around zero [-radius, radius]
x = np.random.rand(NUM_SAMPLES) * (RADIUS * 2) - RADIUS
y = np.random.rand(NUM_SAMPLES) * (RADIUS * 2) - RADIUS
v = (x*x+y*y)

# Display
colors = np.where(v < RADIUS**2, 'green', 'red')
plt.axes().set_aspect('equal', 'box')
# TODO : add a max display ! 
plt.scatter(x, y, color=colors, s=0.5, marker='.')

# Result
total_area = (RADIUS*2)**2
num_green = np.sum(v < RADIUS**2)
print('analytic integration : ', np.pi * RADIUS**2)
print('numerical integration : ', num_green/NUM_SAMPLES*total_area)
```
